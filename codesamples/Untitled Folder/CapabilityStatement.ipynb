```python
import os
import json
import requests

FHIR_BASE = "http://localhost:8080/fhir/"

headers = {"Content-Type": "application/fhir+json;charset=utf-8"}
```

```python
r = requests.get(url = FHIR_BASE + "metadata", headers = headers)
```

```python
print(r.text)
```

```python
cs = json.loads(r.text)
```

```python
print(json.dumps(cs, indent=2))
```

```python
rest = cs["rest"][0]
```

```python
for x in rest:
    resource = x["resource"]
    print(len(resource))
    for r in resource:
        print(r["type"])
    #print(json.dumps(cs["rest"], indent=2))
```
